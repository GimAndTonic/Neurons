```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# Set up the initial plot
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
line, = ax.plot(x, np.sin(x))

def update_plot(frame):
    """Update the plot with a new phase."""
    phase = 2 * np.pi * frame / 20
    line.set_ydata(np.sin(x + phase))

# Create an animation
ani = FuncAnimation(fig, update_plot, frames=20, interval=500, repeat=False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output
import time

# Define your calculation function
def calculate_new_values(phase):
    # Replace this with your actual calculation
    return np.sin(x + phase)

# Set up the initial plot
x = np.linspace(0, 2 * np.pi, 100)

# Main loop for updating the plot
for phase in np.linspace(0, 2 * np.pi, 20):
    new_values = calculate_new_values(phase)

    # Create a new figure for each update
    plt.figure()
    plt.plot(x, new_values)

    # Display the plot
    display(plt.gcf())    
    clear_output(wait=True)  # Clear the output to display the next plot
    time.sleep(0.5)  # Update every half second
```
